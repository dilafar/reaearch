```python
import pandas as pd
import numpy as np
%matplotlib inline
from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf 
from statsmodels.tsa.seasonal import seasonal_decompose      
from pmdarima import auto_arima                              
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv("./cpu_usage.csv", parse_dates=[0])
df = df.set_index('Timestamp').asfreq(freq='5min')


df.drop(columns=[
    'CPU cores',
    'CPU capacity provisioned [MHZ]',
    'CPU usage [MHZ]',
    'Memory capacity provisioned [KB]',
    'Memory usage [KB]',
    'Memory usage [%]',
    'Disk read throughput [KB/s]',
    'Disk write throughput [KB/s]',
    'Disk size [GB]',
    'Network received throughput [KB/s]',
    'Network transmitted throughput [KB/s]'
] , inplace=True)
print(len(df))
print(df.head())
```

```python
title='ARIMA_CPU_USAGE'
ylabel='cpu_usage'
xlabel='' # we don't really need a label here

ax = df['CPU usage [%]'].plot(figsize=(12,5),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
auto_arima(df['CPU usage [%]'],seasonal=True,m=12).summary()
```

```python
train = df.iloc[:len(df)-12]
test = df.iloc[len(df)-12:]
```

```python
model = SARIMAX(train['CPU usage [%]'],order=(0,1,0),seasonal_order=(2,0,0,12))
results = model.fit()
results.summary()
```

```python
start=len(train)
end=len(train)+len(test)-1
predictions = results.predict(start=start, end=end, dynamic=False, typ='levels').rename('SARIMA(0,1,0)(2,0,0,12) Predictions')
```

```python
title='ARIMA CPU Usage'
ylabel='cpu_usage'
xlabel=''

ax = test['CPU usage [%]'].plot(legend=True,figsize=(12,6),title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
error1 = mean_squared_error(test['CPU usage [%]'], predictions)
error2 = rmse(test['CPU usage [%]'], predictions)
print(f'SARIMA(0,1,0)(2,0,0,12) MSE Error: {error1:11.10}')
print(f'SARIMA(0,1,0)(2,0,0,12) RMSE Error: {error2:11.10}')
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(test['CPU usage [%]'], predictions)
print("MAE:", mae)
```
